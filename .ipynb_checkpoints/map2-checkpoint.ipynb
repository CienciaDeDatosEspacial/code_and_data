```python
import pandas as pd
import xarray as xr

# create dataFrame
df = pd.DataFrame({'x':[1,2,3,4],
                   'y':[9.8,19.6,29.4,39.2]}).set_index('x')

# convert to dataset
ds = df.to_xarray()
```

```python
ds
```

```python
ds.info()
```

```python
from datetime import datetime

# change variable names
ds = ds.rename({'x':'time', 'y':'velocity'})

# variable attributes
ds['velocity'].attrs = {'units':'m/s', 'long_name':'free-fall velocity'}
ds['time'].attrs = {'units':'s', 'long_name':'time'}

# global attributes
ds.attrs = {'creation_date':datetime.now(), 'author':'Jane Doe', 'email':'address@email.com'}
```

```python
ds.info()
```

```python

from datetime import datetime

# define data with variable attributes
data_vars = {'velocity':(['time'], [9.8,19.6,29.4,39.2], 
                         {'units': 'm/s', 
                          'long_name':'free-fall velocity'})}

# define coordinates
coords = {'time': (['time'], [1,2,3,4])}

# define global attributes
attrs = {'creation_date':datetime.now(), 
         'author':'Jane Doe', 
         'email':'address@email.com'}

# create dataset
ds = xr.Dataset(data_vars=data_vars, 
                coords=coords, 
                attrs=attrs)
```

```python
ds
```

```python
import numpy as np
import xarray as xr

# create data
data = np.random.randn(2, 3)

# create coords
rows = [1,2]
cols = [1,2,3]

# put data into a dataset
ds = xr.Dataset(
    data_vars=dict(
        variable=(["x", "y"], data)
    ),
    coords=dict(
        lon=(["x"], rows),
        lat=(["y"], cols),
    ),
    attrs=dict(description="coords with vectors"),
)
```

```python
# create data
data = np.random.randn(2, 3, 3)

# create coordinates
rows = [1,2]
cols = [1,2,3]
row_meshgrid, col_meshgrid = np.meshgrid(rows, cols, indexing='ij')
time = pd.date_range("2000-01-01", periods=3)

# put data into a dataset
ds = xr.Dataset(
    data_vars=dict(
        variable=(["x","y","time"], data, {"units":"m/s"})
    ),
    coords=dict(
        row=(["x","y"], row_meshgrid),
        col=(["x","y"], col_meshgrid),
        time=(["time"], time)
    ),
    attrs=dict(description="coords with matrices"),
)
```
